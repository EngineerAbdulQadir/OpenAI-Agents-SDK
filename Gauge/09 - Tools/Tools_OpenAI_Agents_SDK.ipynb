```python
from agents import enable_verbose_stdout_logging, function_tool

enable_verbose_stdout_logging()

@function_tool
def get_current_weather(city:str) -> str:
    """
    Returns Weather For The City

    Args:
        city (str): City Name

    Returns:
        str: Weather

    """
    return f"The weather in {city} is sunny"

@function_tool
def get_current_news(industry:str) -> str:
    """
    Returns Current News For A Particular Industry

    Args:
        industry: The industry to look for the latest news.

    Returns:
        news

    """
    return f"{industry} have new concept of 'Agent Native Cloud'"

agent: Agent = Agent(name="Assistant",
                     instructions="You are a helpful assistant",
                     model="gemini-2.0-flash",
                     tools=[get_current_weather, get_current_news])

result = Runner.run_sync(agent, "What is the weather in Karachi and share latest news in Tech?")

print(result.final_output)
```

```python
from agents import enable_verbose_stdout_logging, function_tool

enable_verbose_stdout_logging()

@function_tool
def get_current_weather(city:str) -> str:
    """
    Returns Weather For The City

    Args:
        city (str): City Name

    Returns:
        str: Weather

    """
    return f"The weather in {city} is sunny"

@function_tool
def get_current_news(industry:str) -> str:
    """
    Returns Current News For A Particular Industry

    Args:
        industry: The industry to look for the latest news.

    Returns:
        news

    """
    return f"{industry} have new concept of 'Agent Native Cloud'"

xactrix_expert: Agent = Agent(name="Xactrix Assistant",
                     instructions="You are a helpful xactrix assistant. Your work is to answer queries that are relevant to Xactrix AI.",
                     model="gemini-2.0-flash",
                     handoff_description="Xactrix Expert")

gloprog_expert: Agent = Agent(name="GloProg Assistant",
                     instructions="You are a helpful gloprog assistant. Your work is to answer queries that are relevant to GloProg.",
                     model="gemini-2.0-flash",
                     handoff_description="GloProg Expert")

triage_agent: Agent = Agent(name="Triage",
                     instructions="You are general chat assistant. Your work is here to observe interaction between the user and let relevant agents answer the queries for the user.",
                     model="gemini-2.0-flash",
                     tools=[get_current_weather, get_current_news],
                     handoffs=[xactrix_expert, gloprog_expert])


result = Runner.run_sync(triage_agent, "Who is the founder of Xactrix AI")

print(result.final_output)
```
